```python
import pandas as pd
import numpy as np

data = pd.read_csv("multiple_linear_regression_dataset.csv")

print(data.head())
print(data.columns)
print(data.shape)

# THINK ABOUT THIS:
# 1. Which columns are inputs?
#    Answer: The inputs are "Years Experience" and "TestScore".
#
# 2. Which column is the output?
#    Answer: The output (target) is "Salary".
#
# 3. How many features does your model need to handle?
#    Answer: It needs to handle 2 features (Experience and TestScore).
```

```python
X = data[["age", "experience"]]

y = data["income"]

# THINK ABOUT THIS:
# 1. What is the shape of x?
#    Answer: The shape is (N, 2) because we are using "age" and "experience" as features.
#
# 2. What is the shape of y?
#    Answer: The shape is (N,) because we are predicting the single "income" column.
```

```python
n_features = X.shape[1]

w = np.zeros(n_features)
b = 0.0

# ---------------------------------------------------------
# THINK ABOUT THIS:
# 1. Why do we need one weight per feature?
#    Answer: Each feature affects the outcome differently (e.g., Experience might be more important than TestScore).
#            Separate weights allow the model to learn the specific importance of each.
#
# 2. Why is bias separate?
#    Answer: Bias allows the line to shift up or down. Without it, the model would always predict 0
#            when inputs are 0, which isn't always true (e.g., base salary).
#
# 3. Would initializing with large values be risky?
#    Answer: Yes. Large weights can cause the prediction error to be massive initially, leading to
#            unstable gradients (exploding gradients) and preventing the model from learning.
```

```python
def predict(X, w, b):
    y_hat = X.dot(w) + b
    return y_hat

# THINK ABOUT THIS:
# 1. Why is there no activation function?
#    Answer: This is linear regression. We want to predict a continuous real number (salary),
#            not a probability or a classification category.
#
# 2. What kind of values can y_hat take?
#    Answer: It can take any real number value (negative to positive infinity).
#
# 3. How is this different from logistic regression?
#    Answer: Logistic regression uses a sigmoid function to squash the output between 0 and 1
#            for classification. Linear regression keeps the raw value.
```

```python
def mean_squared_error(y, y_hat):
    loss = ((y_hat - y) ** 2).mean()
    return loss

# ---------------------------------------------------------
# THINK ABOUT THIS:
# 1. Why square the error?
#    Answer: Squaring ensures negative errors don't cancel out positive ones, and it penalizes
#            large errors more heavily than small ones.
#
# 2. What happens if one prediction is very wrong?
#    Answer: The squared error for that point becomes huge, dominating the loss and forcing the
#            model to adjust significantly to fix it.
#
# 3. Why not just take absolute error?
#    Answer: Absolute error (MAE) is harder to optimize mathematically because its derivative
#            is undefined at 0 (it has a sharp corner), unlike the smooth curve of squared error.
```

```python
def compute_gradients(X, y, y_hat):
    N = len(y)
    dw = (2/N) * X.T.dot(y_hat - y)
    db = (2/N) * (y_hat - y).sum()
    return dw, db

# ---------------------------------------------------------
# THINK ABOUT THIS:
# 1. Why does x appear in dw but not in db?
#    Answer: Because in the equation y = wx + b, w is multiplied by x. When taking the derivative,
#            the x term remains associated with w. Bias b is just added, so its derivative doesn't involve x.
#
# 2. Why does the error term appear everywhere?
#    Answer: The error (y_hat - y) determines the direction and size of the adjustment.
#            If error is 0, we don't need to change anything.
#
# 3. What happens if error is zero?
#    Answer: If error is zero, the gradients become zero, and the weights stop updating (convergence).
```

```python
def update_parameters(w, b, dw, db, lr):
    w = w - lr * dw
    b = b - lr * db
    return w, b
```

```python
lr = 0.0001
epochs = 1000

for epoch in range(epochs):
    y_hat = predict(X, w, b)
    loss = mean_squared_error(y, y_hat)
    dw, db = compute_gradients(X, y, y_hat)
    w, b = update_parameters(w, b, dw, db, lr)
    
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

# ---------------------------------------------------------
# THINK ABOUT THIS:
# 1. Does loss decrease over time?
#    Answer: Yes, it should decrease rapidly at first and then level off as it finds the optimal weights.
#
# 2. What happens if it increases?
#    Answer: If loss increases, the learning rate (lr) is likely too high, causing the model to
#            overshoot the minimum.
#
# 3. How do learning rate and epochs interact?
#    Answer: Lower learning rate requires more epochs to converge. Higher learning rate needs fewer
#            epochs but risks instability/overshooting.
```

```python
print("Final weights:", w)
print("Final bias:", b)

# Example: Predicting income for a person who is 35 years old with 6 years of experience
# format: [age, experience]
new_candidate = np.array([35, 6])

predicted_income = new_candidate.dot(w) + b
print("Predicted Income:", predicted_income)

# ---------------------------------------------------------
# THINK ABOUT THIS:
# 1. Is the prediction reasonable?
#    Answer: Compare the output to the "income" column in your CSV.
#            (e.g., A 30-year-old with 3 years exp earns ~35k in your data, so a 35-year-old should earn more).
```
